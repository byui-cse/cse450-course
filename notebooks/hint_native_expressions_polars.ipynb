```
import polars as pl

# Create a data frame to test with
data = pl.DataFrame({"celsius": [10,20,30,40], "weather": ['sun', 'sun', 'rain', 'rain']})
data
```

```
# Use native expressions API to perform an operation on the entire celsius column
data = data.with_columns(((pl.col('celsius') * 9 / 5) + 32).alias('fahrenheit'))

data
```

```
# Even complicated functions that operate on a row can be done calculated using
# the native expressions API:

# Example function:
def should_go_to_beach(row):
  # We will only go to the beach if it's sunny and warm
  if row['weather'] == 'sun' and row['celsius'] >= 20:
    return True
  else:
    return False

# Native expressions api:
data = data.with_columns(
    ((pl.col('weather') == 'sun' ) & (pl.col('celsius') >= 20)).alias('beach_trip')
)
data
```
